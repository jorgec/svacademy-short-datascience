# Display the countries in ASEAN
Since `ISO country code` isn't the index anymore, selecting certain rows based on a list can be tricky. We've seen how we can select rows using comparison operators (i.e. `<`, `>`, `<=`, `>=`, `==`, `!=`), but those only evaluate single values. To select using a list, we use `isin()`:

```python
asean_iso = ['LAO', 'IDN', 'KHM', 'MMR', 'PHL', 'THA', 'VNM']
df[df['ISO country code'].isin(asean_iso)]
```

# Display the `Sub-national region` with the highest `Intensity of deprivation Regional` in the Philippines:

```python
df[(df['Country'] == 'Philippines')].sort_values('Intensity of deprivation Regional', ascending=False).head(1)
```

## And the top for each country in the ASEAN:

```python
df[df['ISO country code'].isin(asean_iso)].sort_values('Intensity of deprivation Regional', ascending=False).drop_duplicates('Country')
```
