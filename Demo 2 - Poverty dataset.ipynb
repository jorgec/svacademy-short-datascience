## Description of the data
- **ISO**: Unique ID for country
- **Country**: country name
- **MPI Urban**: Multi-dimensional poverty index for urban areas within the country
- **Headcount Ratio Urban**: Poverty headcount ratio (% of population listed as poor) within urban areas within the country
- **Intensity of Deprivation Urban**: Average distance below the poverty line of those listed as poor in urban areas
- **MPI Rural**: Multi-dimensional poverty index for rural areas within the country
- **Headcount Ratio Rural**: Poverty headcount ratio (% of population listed as poor) within rural areas within the country
- **Intensity of Deprivation Rural**: Average distance below the poverty line of those listed as poor in rural areas

This is a publicly available dataset provided by the Oxford Poverty & Human Development Initiative via [Kaggle](https://www.kaggle.com/ophi/mpi)

## By Row

### .loc[]
.loc is primarily label based, but may also be used with a boolean array. .loc will raise KeyError when the items are not found. Allowed inputs are:
- A single label, e.g. 5 or 'a' (Note that 5 is interpreted as a label of the index. This use is not an integer position along the index.).
- A list or array of labels ['a', 'b', 'c'].
- A slice object with labels 'a':'f' (Note that contrary to usual python slices, both the start and the stop are included, when present in the index! See Slicing with labels.).
- A boolean array
- A callable function with one argument (the calling Series, DataFrame or Panel) and that returns valid output for indexing (one of the above).

To properly use `.loc[]`, we need to specify an __index__; let's use `ISO`:

```python
df.set_index('ISO', inplace=True)
```

Now we can use the `ISO` code to search for a specific row; for example, the Philippines, which has an `ISO` code of 'PHL':

```python
df.loc['PHL']
```

You can also select multiples:

```python
df.loc[['PHL', 'SOM']]
```

Let's create a list of `ISO` codes for ASEAN: ['LAO', 'IDN', 'KHM', 'MMR', 'PHL', 'THA', 'VNM'], excluding Brunei, Malaysia, and Singapore, as they aren't included in the dataset:

```python
asean_iso = ['LAO', 'IDN', 'KHM', 'MMR', 'PHL', 'THA', 'VNM']
```

```python
df.loc[asean_iso]
```

Let's now compare Headcount Rations (Urban and Rural) across the ASEAN countries:

```python
df.loc[asean_iso][['Headcount Ratio Urban', 'Headcount Ratio Rural']].plot(kind='bar', figsize=(30,20))
```

### Using Filters

We can filter the rows by passing a condition that gets evaluated for every row; for example, to see all countries with a `Headcount Ratio Urban` greater than 50, we start by first evaluating the statement `df['Headcount Ratio Urban'] > 50`:

```python
df['Headcount Ratio Urban'] > 50
```

...which returns a series of Boolean (True/False) values per row. We then pass that statement to the dataframe, like so: `df[df['Headcount Ratio Urban'] > 50`].

For the sake of readability, we can put that inside a variable:

`headcount_filter = df['Headcount Ratio Urban'] > 50`

... and then pass that variable to the dataframe:

`df[headcount_filter]`

```python
headcount_filter = df['Headcount Ratio Urban'] > 50
df[headcount_filter]
```

We can chain these filters using the Boolean operators `&` and `|`:

```python
headcount_filter = df['Headcount Ratio Urban'] > 30
deprivation_filter = df['Intensity of Deprivation Urban'] > 20
df[headcount_filter & deprivation_filter]
```
